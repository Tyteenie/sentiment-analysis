```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')
```

```python
start_date = pd.to_datetime('2022-11-01')
df['Day'] = pd.to_timedelta(df['Day'] - 1, unit='D') + start_date
df.columns = ['date', 'Revenue']
df.set_index('date', inplace=True)

# Display the DataFrame
df
```

```python
#plotting time series
df['Revenue'].plot(figsize=(12,5))
plt.title('Revenue in Millions')
plt.xlabel('Date')
plt.ylabel('Revenue in million dollars')
plt.grid(True)
plt.show()
```

```python
# Plot the rolling mean to check for stationarity
df.rolling(window=30).mean().plot()
```

```python
#export cleaned data set
pd.DataFrame(df).to_csv("cleanedData.csv")
```

```python
# Conduct an Augmented Dickey-Fuller test on the data to assess stationarity.
num_observations = len(df['Revenue'])
result = adfuller(df['Revenue'], autolag='AIC')
# Print the results
print("Number of observations: ", num_observations)
print("Test statistics: ", result[0])
print("p-value: ", result[1])
print("Critical values (Tab-separated):")
for key, value in result[4].items():
    print("\t",key, ": ",value)
if result[1] <= 0.05:
    print("Reject Null Hypothesis, The Time Series is stationary")
else:
    print("Fail to reject Null Hypothesis, The Time Series is non-stationary")
```

```python
#make the time series stationary
dfstat = df.diff().dropna()
dfstat.head(5)
```

```python
# Perform Augmented Dicky-Fuller on data to test for stationarity again
num_observations = len(dfstat['Revenue'])
result = adfuller(dfstat['Revenue'], autolag='AIC')
# Print the results
print("Number of observations: ", num_observations)
print("Test statistics: ", result[0])
print("p-value: ", result[1])
print("Critical values (Tab-separated):")
for key, value in result[4].items():
    print("\t",key, ": ",value)
if result[1] <= 0.05:
    print("Reject Null Hypothesis, The Time Series is stationary")
else:
    print("Fail to reject Null Hypothesis, The Time Series is non-stationary")
```

```python
# Plot to verify stationarity
dfstat.plot();
```

```python
# Plot the differenced rolling mean to check for stationarity
dfstat['Revenue'].rolling(window=30).mean().plot()
```

```python
# Decompose the transformed data
#perform decomposition
decomp = seasonal_decompose(dfstat['Revenue'], period=90)
# Plot the decomposition
decomp.plot()
plt.gcf().set_size_inches(8, 6)
plt.suptitle('Revenue', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#Seasonality
plt.title('Seasonality')
decomp.seasonal.plot()
plt.gcf().set_size_inches(6, 3)
```

```python
#Trend
plt.title('Trend')
decomp.trend.plot()
plt.gcf().set_size_inches(6, 3)
```

```python
#Residuals
plt.title('Residuals')
decomp.resid.plot()
plt.gcf().set_size_inches(6, 3)
```

```python
# Generate predictions and plot to line graph with test and training data
from statsmodels.graphics.tsaplots import plot_predict

def arima_forecast_plot(model, x_train, x_test, title, ylabel, xlabel, start=len(x_train), end=len(x_train)+len(x_test), figsize=(16, 8)):
    fig, ax = plt.subplots(figsize=figsize)
    ax.set(title=title, xlabel=xlabel, ylabel=ylabel)
    x_train.plot(ax=ax)
    x_test.plot(ax=ax, color='teal')
    plot_predict(model, start=start, end=end, ax=ax, alpha=0.05)
    ax.annotate('Predictions Start', xy=(x_train.index[-1], x_train.iloc[-1]), xytext=(x_train.index[-1], x_train.iloc[-1] + 0.5), arrowprops=dict(facecolor='black', shrink=0.1))
    ax.legend(['Training Data', 'Testing Data', 'Predicted Values'])
    plt.show()
    
arima_forecast_plot(res, x_train, x_test, title='Daily Revenue Time Series', ylabel='Daily Revenue (in millions)', xlabel='Time Horizon')
```
